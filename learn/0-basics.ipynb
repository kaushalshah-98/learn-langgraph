## What is Lang graph
-----------------------

It is a library for building stateful, multi factor applications with LLM's used to create agent and multi agent workflows. Compared to other LLM frameworks, it offer three core benefits
- Cycles
- Controllability
- Persistence
It allows you to define flows that involves cycles, essentials for multi agent architectures, differentiating it from DAG-based solutions.

## Why Lang graph
-----------------------
1. It simplifies development
   1. State management of agents
   2. Agent coordination
   3. Conditional logics
   4. Complex work is split into multi agent work, with one agent working on specific task
   
2. Flexibility
   1. We can define our own agent logic and communication protocol.
   2. This allows for highly customized applications tailored to specific use cases.
   3. Whether we need a chatbot that can handle various types of user requests or multi agent system that performs complex tasks.
   4. It provides the tool to build exactly what we need, it has all the power to create.
   
3. Scalability
   1. Large scale multi agent applications
   2. handle high volume of interactions, interaction can be btw agents and complex workflow
   3. It also offers enterprise level application

4. Fault tolerance
   1. Handle errors 
   2. if one agent fails, we have fault tolerance mechanism to keep it running


## State

- It is shared data structure that holds the current information or context of entire application
- Its like application memory, keeping track of variables and data tha node can access and modify as they execute
- eg whiteboard (state) in meeting room where peoples (nodes) can write and read info and coordinate

## Nodes

- Individual functions or operations that perform specific tasks in a graph
- Each node receives an input as state and it produce output or an updated state
- eg assembly line, each station does one job, attach a part, paint it, inspect and so on

## Graph

- It is overarching structure  that maps out how different task (nodes) are connected and executed
- Visually represent workflow, showing sequence and conditional paths btw various operations
- eg roadmap, displaying routes connecting cities with intersections offering choices on which path to go next

## Edges

- They are connections btw nodes that determine flow of execution
- They tells which node to execute next after current one completes its task
- eg train track- connect stations (nodes) in specific direction.

## Conditional edge

- They are specialize connections that decide next node to execute based on specific conditions or logic applied to current state
- eg traffic light- each light has different actions to decide next step

## Start

- Virtual node entry point in graph where workflow begins
- It does not perform any operations but serve as start position in graph
- eg race starting line, place where race begins

## End

- Virtual node end point in graph where workflow ends
- upon reaching, the graph executions stops indicating that all process have been completed
- eg finish line, place where race ends 

## Tools

- They are specialized functions or utils that nodes can utilize to perform operations such as fetching and data from API
- They enhance capability of nodes by providing extra functionality and offload workload
- Nodes are part of graph structure while tools are functionality used within nodes
- eg tools in a toolbox, hammer for nails, screwdriver for screw, each tool has some purpose

## Toolnode

- Special kind of node whose main job is to run a tool.
- It connects tools output back into state so other nodes can use that information.
- eg operator (toolnode) controls machine (tool), then take result to assembly line

## Stategraph

- Class to build and compile graph structure
- It manage node,edges and overall state ensuring workflow operates in unified way and data flow correctly btw components
- eg blueprint of building outlines design and connections within

## Runnable

- Its standardized executable component that performs a specific task within AI workflow
- It serves as fundamental building block, allowing for us to create modular systems
- eg lego bricks can be snapped together to build complex structure, runnable can be combined to created AI workflows

## Messages type

1. Human message - Input from a user
2. AI message - Response generated by AI models
3. System message - Used to provide info or context to AI model
4. Tool message - Similar to func message but specific to tool usage
5. Function message - Result of functional call


```python
## Hello world graph Objective
'''
1. Understand and define agent state structure
2. Create simple node functions to process and update state
3. Setup langgraph structure
4. Compile and invoke a graph
5. Understand data flows though nodes
6. Start -> Node -> End
'''

from typing import TypedDict
from langgraph.graph import StateGraph, START, END
from IPython.display import Image, display

class AgentState(TypedDict):
    name: str
    message: str

def greet(state: AgentState) -> AgentState:
    nameInput = state.get('name', 'Unknown')
    return {"message": f"Hey {nameInput}, How are you"}

graph = StateGraph(AgentState)
graph.add_node('greet', greet)
graph.set_entry_point('greet')
graph.set_entry_point('greet')

app = graph.compile()

# display(Image(app.get_graph().draw_mermaid_png()))

result = app.invoke({'name': "Kaushal"})
result
```

```python
## Multiple inputs graph Objectives
'''
1. Define complex Agentstate
2. Create node that perform operations on list data
3. Setup langgraph structure
4. Invoke graph and retrieve outputs

Main goal:- to handle multiple inputs
'''

from typing import TypedDict, List
from langgraph.graph import StateGraph, START, END
from IPython.display import Image, display

class AgentState(TypedDict):
    values: List[int]
    name: str
    result: str

def process_val(state: AgentState) -> AgentState:
    nameInput = state.get('name', 'Unknown')
    sumData = sum(state.get('values', []))
    result = f"Hey {nameInput}, How are you, your sum i {sumData}"
    state['result'] = result
    # return { "result": result }
    return state

graph = StateGraph[AgentState, None, AgentState, AgentState](AgentState)
graph.add_node('processor', process_val)
graph.set_entry_point('processor')
graph.set_entry_point('processor')

app = graph.compile()

# display(Image(app.get_graph().draw_mermaid_png()))

result = app.invoke({'values': [1,2], "name": "Kaushal"})
result['result']
```

```python
## Sequential Graph Objectives
'''
1. Create multiple nodes that sequential process and update diff parts of state
2. Connect nodes in graph
3. Invoke graph and see how state is transformed by nodes

Goal:= Create and handle multiple nodes
'''

from typing import TypedDict, List
from langgraph.graph import StateGraph, START, END
from IPython.display import Image, display

class AgentState(TypedDict):
    age: int
    name: str
    final: str

def first_node(state: AgentState) -> AgentState:
    nameInput = state.get('name', 'Unknown')
    state['final'] = f"Hey {nameInput}, How are you,"
    return state

def second_node(state: AgentState) -> AgentState:
    ageInput = state.get('age', '0')
    state['final'] += f"You are {ageInput} years old"
    return state

graph = StateGraph[AgentState, None, AgentState, AgentState](AgentState)
graph.add_node('first_node', first_node)
graph.add_node('second_node', second_node)

# graph.add_edge(START, 'first_node')
graph.set_entry_point('first_node')
graph.add_edge('first_node', 'second_node')
graph.set_finish_point('second_node')
# graph.add_edge('second_node', END)

app = graph.compile()

# display(Image(app.get_graph().draw_mermaid_png()))

result = app.invoke({"name": "Kaushal", "age": 42})
result['final']
```

```python
## Conditional Graph Objectives
'''
1. Implement conditional logic to route flow of data to different node
2. Use start and end nodes to manage entry/exit
3. Design nodes to perform different operations (addition, subtraction)
4. Create router node to handle decision making and control graph flow

Goal:= How to use add_conditional_edges()
'''

from typing import TypedDict, List
from langgraph.graph import StateGraph, START, END
from IPython.display import Image, display

class AgentState(TypedDict):
    operation: int
    number1: int
    number2: int
    final: int

def add(state: AgentState) -> AgentState:
    state['final'] = state.get('number1', 0) + state.get('number2', 0)
    return state

def sub(state: AgentState) -> AgentState:
    state['final'] = state.get('number1', 0) - state.get('number2', 0)
    return state

def decide(state: AgentState) -> AgentState:
    if state.get('operation') == '+':
        return 'add'
    return 'sub'

graph = StateGraph[AgentState, None, AgentState, AgentState](AgentState)
graph.add_node('add', add)
graph.add_node('sub', sub)
graph.add_node('router', lambda x: x)

graph.add_edge(START, 'router')
graph.add_conditional_edges('router', decide, {'add': 'add', 'sub': 'sub'})
graph.add_edge('add', END)
graph.add_edge('sub', END)

app = graph.compile()

# display(Image(app.get_graph().draw_mermaid_png()))

result = app.invoke({"number1": 2, "number2": 2, "operation": "+"})
result['final']
```

```python
## Looping Graph Objectives
'''
1. Implement looping logic to route the data flow back to nodes
2. Create single conditional edge to handle decision making and control graph

Goal:= Looping logic
'''

from typing import TypedDict, List
from langgraph.graph import StateGraph, START, END
from IPython.display import Image, display

class AgentState(TypedDict):
    name: int
    loopCount: int
    count: int
    final: str

def greet2(state: AgentState) -> AgentState:
    state['final'] = f"Hey {state.get('name', 'stranger')}!"
    print('final',state['final'])
    state['count'] = 0
    return state

def looper(state: AgentState) -> AgentState:
    state['final'] += f", We had loop {state.get('count')}  times!"
    # print('count',state['count'])
    state['count']+=1
    return state
    # return {"count": state["count"] + 1}

def decide2(state: AgentState) -> AgentState:
    if state.get('count') < state.get('loopCount'):
        return 'loop'
    return 'end'

graph1 = StateGraph[AgentState, None, AgentState, AgentState](AgentState)
graph1.add_node('greet2', greet2)
graph1.add_node('looper', looper)

# graph.add_edge(START, 'greet')
graph1.set_entry_point('greet2')
graph1.add_edge('greet2', 'looper')
graph1.add_conditional_edges('looper', decide2, {'loop': 'looper', 'end': END})
# graph.add_edge('looper', END)

app1 = graph1.compile()

# display(Image(app.get_graph().draw_mermaid_png()))

results = app1.invoke({"name": "Kaushal", "loopCount": 4, "count": 0})
results['final']
```

```python
## Tool

from typing import Annotated, Sequence, TypedDict
from langgraph.graph import StateGraph, START, END
from langchain_ollama import ChatOllama
from langchain_core.messages import BaseMessage, SystemMessage, ToolMessage
from langgraph.graph.message import add_messages
# from langchain_core.messages import HumanMessage, AIMessage
from langchain_core.tools import tool
from langgraph.prebuilt import ToolNode

## reducer function
## rule that controls how updates from node are combined with existing state
## tells us how to merge new data into current state
## without reducer, update would have replaced existing value entirely!

class State(TypedDict):
    messages: Annotated[Sequence[BaseMessage], add_messages]

@tool
def add(a: int, b:int):
    """ This is an add function that add two numbers"""
    return a + b

@tool
def sub(a: int, b:int):
    """ This is an add function that add two numbers"""
    return a - b

@tool
def multiply(a: int, b:int):
    """ This is an add function that add two numbers"""
    return a * b

tools = [add,sub,multiply]
model = llm.bind_tools(tools)

def model_call(state: State) -> State:
    systemPrompt = SystemMessage(content = 'You are AI assistant. Answer query to best of your ability')
    response = model.invoke([systemPrompt] + state['messages'])
    return {"messages": [response]}

def should_continue(state: State):
    last_message = state['messages'][-1]
    if not last_message.tool_calls:
        return 'end'
    else:
        return 'continue'

graph = StateGraph(State)
graph.add_node('agent', model_call)
tool_node = ToolNode(tools=tools)
graph.add_node('tools', tool_node)

graph.set_entry_point('agent')
graph.add_conditional_edges('agent', should_continue, {'continue': 'tools', 'end': END})
graph.add_edge('tools', 'agent')

app = graph.compile()

def print_stream(stream):
    for s in stream:
        message = s['messages'][-1]
        if isinstance(message, tuple):
            print(message)
        else:
            message.pretty_print()

inputs = {"messages": [
      {"role": "user", "content": "Add 4 and 5 and multiply by 2"}
    ]}
print_stream(app.stream(inputs, stream_mode='values'))
```
